# 1 - A NASTY INTEGRAL

```python
rng = np.random.default_rng()
```

```python
def f(x):
    y = x**3
    return abs(y)
```

```python
sigma = 2
N = 100000

x = rng.normal(scale = sigma, size = N)

integ = np.mean(f(x)) * np.sqrt(2 * np.pi) * sigma / 2
print(integ)
```
