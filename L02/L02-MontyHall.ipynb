```python
doors = [0, 1, 2]

N = 1000

n_s = 0
n_c = 0
n_n = 0

for i in range(N):
    win = np.random.choice(doors, 1)
    cons = np.random.choice(doors, 1)
    swit = np.random.choice(doors, 1)
    newc = np.random.choice(doors, 1)
    goat = [0, 1, 2]
    if swit == win:
        goat.remove(win)
        swit = np.random.choice(goat, 1)
    else:
        swit = win

    if swit == win:
        n_s += 1

    if cons == win:
        n_c += 1
        
    if newc == win:
        n_n += 1
```

```python
print(n_s / N, n_c / N, n_n / N)
```
