## Tasks

- Simulate three doors, one car and two goats.
- Simulate three players: the switcher, the conservative and the newcomer.
- Record who wins.
- Repeat it many times.
- Which player do you want to be?

Let's think about this.

Whatever the conservative does he's stuck with his choice. So I just need to check if he won or not.

If the swithcer chooses the winning door then it changes it, hence he loses. If he chooses a goat, the other one is revealed and he will change to the car. So, if his initial choice is a losing one, he end up winning.

Put in these terms, it's evident that the swithcer has double the chance to win.

```python
doors = [0, 1, 2]

N = 10000

n_s = 0
n_c = 0
n_n = 0

for i in range(N):
    win = np.random.choice(doors)
    cons = np.random.choice(doors)
    swit = np.random.choice(doors)
    newc = np.random.choice([win, 3]) # 3 just to be sure the choice is between the winning door and something different

    if swit != win:
        n_s += 1

    if cons == win:
        n_c += 1
        
    if newc == win:
        n_n += 1
```

```python
print(' --- Results ---\n')

print('Switcher win ratio:\t', n_s / N)
print('Conservative win ratio:\t', n_c / N)
print('Newcomer win ratio:\t', n_n / N)
```
